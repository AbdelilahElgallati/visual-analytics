```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import numpy as np
```

```python
# charger l'image
img = Image.open("images/Lena.png")
r, g, b = img.split()
```

```python
# afficher les images
plt.figure(figsize=(10, 5))
plt.hist(np.array(r).ravel(), bins=256, color='red', alpha=0.5, label='Rouge')
plt.hist(np.array(g).ravel(), bins=256, color='green', alpha=0.5, label='Vert')
plt.hist(np.array(b).ravel(), bins=256, color='blue', alpha=0.5, label='Bleu')
plt.title("Histogramme des canaux de couleur")
plt.xlabel("Niveau d'intensité")
plt.ylabel("Nombre de pixels")
plt.legend()
plt.show()
```

```python
# chargement d'image
img2 = np.array(Image.open("images/Lena.png"))
# pour avoir l'histogramme cumulé normalisé: argument cumulative=True, density=True
n, bins, patches = plt.hist(img2.flatten(), bins = range(256))
plt.title("Histogramme d'image")
plt.xlabel("Niveau d'intensité")
plt.ylabel("Nombre de pixels")
plt.show()
```

## auto constrast

```python
img3 = Image.open("images/Lena.png")
img3_auto_contrast = ImageOps.autocontrast(img3)
print("image originale")
display(img3)
print("image auto constrast")
display(img3_auto_contrast)
```

## égalisation d'histogramme avec python

```python
img4 = Image.open("images/Lena.png")
img4_eq = ImageOps.equalize(img4)
print("image originale")
display(img4)
print("image equalize")
display(img4_eq)
```
