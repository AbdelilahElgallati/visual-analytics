```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import numpy as np
```

```python
# charge d'images
img_surexp = Image.open("images/simba_surexp.png")
```

```python
# appliquer les transformations
img_surexp_auto = ImageOps.autocontrast(img_surexp)
img_surexp_eq = ImageOps.equalize(img_surexp)
```

```python
# Afficher les histogrammes
def plot_histogram(image, title):
    plt.figure(figsize=(8, 4))
    plt.hist(np.array(image).ravel(), bins=256, color='black', alpha=0.7)
    plt.title(title)
    plt.xlabel('Niveau de gris')
    plt.ylabel('Nombre de pixels')
    plt.show()
```

```python

plot_histogram(img_surexp, "Histogramme de l'image d'origine")
plot_histogram(img_surexp_auto, "Histogramme de l'image après autocontraste")
plot_histogram(img_surexp_eq, "Histogramme de l'image après égalisation")
```
