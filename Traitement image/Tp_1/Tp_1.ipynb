## Activité 3: traitement d'image

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt

# Charger l'image
image_simba = Image.open("./IMGS/2.png")
```

```python
# Convertir en niveaux de gris
im_gray = ImageOps.grayscale(image_simba)
```

```python
# Fonction pour afficher l'image et son histogramme
def afficher_image_histogramme(image, title):
    plt.figure(figsize=(10, 4))

    # Affichage de l'image
    plt.subplot(1, 2, 1)
    plt.imshow(image, cmap='gray')
    plt.title(title)
    plt.axis("off")

    # Affichage de l'histogramme
    plt.subplot(1, 2, 2)
    plt.hist(image.getdata(), bins=256, range=(0, 256), color='gray', alpha=0.7)
    plt.title("Histogramme")
    plt.xlabel("Intensité des pixels")
    plt.ylabel("Nombre de pixels")

    plt.show()
```

```python
# Ajustement du contraste avec autocontrast()
im_autocontrast = ImageOps.autocontrast(im_gray)
afficher_image_histogramme(im_gray, "Image originale")
afficher_image_histogramme(im_autocontrast, "Autocontrast")
```

```python
# Égalisation de l'histogramme
im_equalized = ImageOps.equalize(im_gray)
afficher_image_histogramme(im_equalized, "Égalisation de l'histogramme")
```

```python
# Inversion des couleurs
im_inverted = ImageOps.invert(im_gray)
afficher_image_histogramme(im_inverted, "Inversion des couleurs")
```

```python
# Effet miroir (flip horizontal)
im_mirror = ImageOps.mirror(im_gray)
afficher_image_histogramme(im_mirror, "Effet miroir (flip horizontal)")
```
