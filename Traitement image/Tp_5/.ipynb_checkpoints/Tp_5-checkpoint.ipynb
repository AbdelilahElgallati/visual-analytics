# TP5 : Classification d'Images avec Deep Learning

Ce notebook implémente les activités du TP5 pour développer, entraîner et évaluer un modèle de Deep Learning pour la classification d'images en 3 classes, ainsi qu'explorer le Transfer Learning avec VGG19.

## Activité 1 : Préparation des données

Nous allons charger et prétraiter les images (redimensionnement, normalisation) et préparer les ensembles d'entraînement, validation et test à l'aide de `ImageDataGenerator`.

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os

# Chemins des datasets
train_dir = 'dataset/train'
validation_dir = 'dataset/validation'
test_dir = 'dataset/test'

# Paramètres
img_height, img_width = 128, 128  # Taille pour le CNN personnalisé
batch_size = 32

# Générateurs de données avec normalisation
train_datagen = ImageDataGenerator(
    rescale=1./255,  # Normalisation
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)

validation_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Chargement des données
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical'
)

validation_generator = validation_datagen.flow_from_directory(
    validation_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Afficher les informations
print(f'Classes détectées : {train_generator.class_indices}')
print(f'Images d\'entraînement : {train_generator.samples}')
print(f'Images de validation : {validation_generator.samples}')
print(f'Images de test : {test_generator.samples}')
```

## Activité 2 : Création et entraînement d'un modèle CNN personnalisé

Nous allons créer un CNN avec l'architecture proposée, le compiler et l'entraîner.

```python
from tensorflow.keras import models, layers

# Création du modèle CNN
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(3, activation='softmax')
])

# Compilation du modèle
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.CategoricalCrossentropy(),
    metrics=['accuracy']
)

# Résumé du modèle
model.summary()

# Entraînement du modèle
history = model.fit(
    train_generator,
    epochs=10,
    batch_size=batch_size,
    validation_data=validation_generator
)
```

## Activité 3 : Évaluation du modèle CNN

Nous allons tracer les courbes de précision et de perte, générer un rapport de classification et afficher une matrice de confusion.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np

# Tracer les courbes de précision et de perte
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(10)

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Précision Entraînement')
plt.plot(epochs_range, val_acc, label='Précision Validation')
plt.title('Précision Entraînement et Validation')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Perte Entraînement')
plt.plot(epochs_range, val_loss, label='Perte Validation')
plt.title('Perte Entraînement et Validation')
plt.legend()

plt.show()

# Évaluation sur l'ensemble de test
test_generator.reset()
predictions = model.predict(test_generator)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = test_generator.classes
class_labels = list(test_generator.class_indices.keys())

# Rapport de classification
print('Rapport de classification :')
print(classification_report(true_classes, predicted_classes, target_names=class_labels))

# Matrice de confusion
cm = confusion_matrix(true_classes, predicted_classes)
plt.figure(figsize=(8, 6))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Matrice de Confusion')
plt.colorbar()
tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks, class_labels, rotation=45)
plt.yticks(tick_marks, class_labels)
plt.ylabel('Classe Réelle')
plt.xlabel('Classe Prédite')
plt.show()

# Modification des hyperparamètres (exemple : augmenter les neurones dans Dense)
model_modified = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),  # Augmenté de 64 à 128
    layers.Dense(3, activation='softmax')
])

model_modified.compile(
    optimizer='adam',
    loss=tf.keras.losses.CategoricalCrossentropy(),
    metrics=['accuracy']
)

# Entraînement du modèle modifié
history_modified = model_modified.fit(
    train_generator,
    epochs=10,
    batch_size=batch_size,
    validation_data=validation_generator
)

# Analyse des résultats
print('Analyse des hyperparamètres modifiés :')
print('Augmenter le nombre de neurones dans la couche dense (de 64 à 128) peut améliorer la capacité du modèle à capturer des caractéristiques complexes, mais peut aussi augmenter le risque de surapprentissage si le dataset est petit.')
```

## Activité 4 : Transfer Learning avec VGG19

Nous allons utiliser VGG19 pré-entraîné, geler ses couches, ajouter de nouvelles couches et entraîner le modèle.

```python
from tensorflow.keras.applications import VGG19
from tensorflow.keras.models import Model

# Préparer les données pour VGG19 (224x224)
img_height_vgg, img_width_vgg = 224, 224

train_datagen_vgg = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True
)

validation_datagen_vgg = ImageDataGenerator(rescale=1./255)
test_datagen_vgg = ImageDataGenerator(rescale=1./255)

train_generator_vgg = train_datagen_vgg.flow_from_directory(
    train_dir,
    target_size=(img_height_vgg, img_width_vgg),
    batch_size=batch_size,
    class_mode='categorical'
)

validation_generator_vgg = validation_datagen_vgg.flow_from_directory(
    validation_dir,
    target_size=(img_height_vgg, img_width_vgg),
    batch_size=batch_size,
    class_mode='categorical'
)

test_generator_vgg = test_datagen_vgg.flow_from_directory(
    test_dir,
    target_size=(img_height_vgg, img_width_vgg),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# Charger VGG19 pré-entraîné
vgg = VGG19(input_shape=(img_height_vgg, img_width_vgg, 3), weights='imagenet', include_top=False)

# Geler les couches de VGG19
for layer in vgg.layers:
    layer.trainable = False

# Ajouter de nouvelles couches
x = layers.Flatten()(vgg.output)
prediction = layers.Dense(3, activation='softmax')(x)
model_vgg = Model(inputs=vgg.input, outputs=prediction)

# Compilation du modèle
model_vgg.compile(
    optimizer='adam',
    loss=tf.keras.losses.CategoricalCrossentropy(),
    metrics=['accuracy']
)

# Résumé du modèle
model_vgg.summary()

# Entraînement du modèle
history_vgg = model_vgg.fit(
    train_generator_vgg,
    epochs=10,
    batch_size=batch_size,
    validation_data=validation_generator_vgg
)

# Évaluation sur l'ensemble de test
test_generator_vgg.reset()
predictions_vgg = model_vgg.predict(test_generator_vgg)
predicted_classes_vgg = np.argmax(predictions_vgg, axis=1)
true_classes_vgg = test_generator_vgg.classes

# Rapport de classification pour VGG19
print('Rapport de classification pour VGG19 :')
print(classification_report(true_classes_vgg, predicted_classes_vgg, target_names=class_labels))

# Matrice de confusion pour VGG19
cm_vgg = confusion_matrix(true_classes_vgg, predicted_classes_vgg)
plt.figure(figsize=(8, 6))
plt.imshow(cm_vgg, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Matrice de Confusion VGG19')
plt.colorbar()
tick_marks = np.arange(len(class_labels))
plt.xticks(tick_marks, class_labels, rotation=45)
plt.yticks(tick_marks, class_labels)
plt.ylabel('Classe Réelle')
plt.xlabel('Classe Prédite')
plt.show()
```

## Comparaison et Analyse

**Comparaison CNN vs VGG19** :
- **CNN Personnalisé** : Ce modèle est plus léger et rapide à entraîner, mais ses performances dépendent fortement de la taille et de la qualité du dataset. Il peut être limité pour des datasets complexes en raison de sa simplicité.
- **VGG19 (Transfer Learning)** : Bénéficie des caractéristiques apprises sur ImageNet, ce qui améliore les performances, surtout avec des datasets plus petits. Cependant, il est plus lourd et nécessite des images en 224x224, augmentant les besoins en calcul.
- **Différences observées** : VGG19 devrait offrir une meilleure précision sur l'ensemble de test grâce à ses caractéristiques pré-entraînées, surtout si le dataset est petit. Le CNN peut surapprendre plus facilement si les données sont limitées.

**Interprétation des résultats** :
- Les courbes de précision et de perte montrent si le modèle apprend efficacement ou s'il surapprend (écart important entre entraînement et validation).
- Le rapport de classification (précision, rappel, F1-score) indique les performances par classe. Une faible performance pour une classe peut suggérer un déséquilibre des données.
- La matrice de confusion révèle les erreurs de classification (par exemple, confusion entre classes similaires).

**Impact des hyperparamètres** : Augmenter les neurones dans la couche dense (de 64 à 128) peut améliorer la précision pour des datasets complexes, mais risque le surapprentissage si les données sont insuffisantes. Ajouter des couches ou modifier l'optimiseur (par exemple, SGD avec élan) pourrait également influencer les performances.
